```python
import pandas as pd
```

```python
courses_df = pd.read_csv("./output/all_courses.csv")
courses_df
```

```python
# what is the course distribution by course type (nanodegree, course, or free)?
course_type_distribution = courses_df.groupby(["Type"]).count()["Title"]
```

```python
course_type_distribution.plot(kind="pie", title="Course Type Breakdown")
```

```python
# what is the course distribution by course level (beginner, intermediate, advanced)?
course_level_distribution = courses_df.groupby(["Level"]).count()["Title"]
```

```python
course_level_distribution.plot(kind="pie", title="Course Level Breakdown")
```

```python
# what is the course distribution by course level and type (beginner, intermediate, advanced)?
course_breakdown_distribution = courses_df.groupby(["Level", "Type"]).count()["Title"]
course_breakdown_distribution
```

```python
courses_df.groupby(['Level', 'Type']).size().unstack().plot(kind='bar', stacked=True, title="Course Type and Level Distribution")
```

```python
# which courses are affiliated with Google?
def hasString(df, col, sample_str):
    result = df[col].str.contains(sample_str)
    result.fillna(value=False, inplace=True)
    return result
    
google_aff_df = courses_df.loc[hasString(courses_df, "Affiliates", "Google")]
print("Google is affiliated with {} Udacity courses!".format(len(google_aff_df)))
google_aff_df
```

```python
# which courses use python?
python_instruction_df = courses_df.loc[hasString(courses_df, "Skills Covered", "Python")]
print("Python is used in {} Udacity courses!".format(len(python_instruction_df)))
python_instruction_df
```

```python
# which courses require python?
python_required_df = courses_df.loc[hasString(courses_df, "Prerequisites", "Python")]
print("Python is required in {} Udacity courses!".format(len(python_required_df)))
python_required_df
```
